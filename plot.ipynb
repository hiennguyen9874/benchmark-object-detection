```python
import matplotlib.pyplot as plt
```

```python
data = [
    ("YOLOv5", "YOLOv5n", 549.38, 0.277, "r"),
    ("YOLOv5", "YOLOv5s", 349.38, 0.371, "r"),
    ("YOLOv5", "YOLOv5m", 194.37, 0.448, "r"),
    ("YOLOV7", "YOLOv7-Tiny", 369.57, 0.37, "g"),
    ("YOLOV7", "YOLOv7", 125.77, 0.508, "g"),
    ("YOLOX", "YOLOX-S", 282.59, 0.403, "b"),
    ("YOLOX", "YOLOX-M", 161.70, 0.468, "b"),
    ("YOLOX", "YOLOX-Nano", 619.27, 0.257, "y"),
    ("YOLOX", "YOLOX-Tiny", 600.67, 0.328, "y"),
]

data_line = (
    ("YOLOv5n", "YOLOv5s", 'r', 'dashdot'),
    ("YOLOv5s", "YOLOv5m", 'r', 'dashdot'),
    ("YOLOv7-Tiny", "YOLOv7", 'g', "dashed"),
    ("YOLOX-S", "YOLOX-M", 'b', "dotted"),
    ("YOLOX-Nano", "YOLOX-Tiny", 'y', "solid"),
)
```

```python
group_data = {x[1]: x for x in data}
```

```python
import itertools
marker = itertools.cycle((',', '+', '.', 'o', '*')) 

for group_name, model_name, fps, map, color in data:
    plt.scatter(fps, map, label=model_name, color=color, marker=next(marker))

for from_name, to_name, color, linestyle in data_line:
    plt.plot((group_data[from_name][2], group_data[to_name][2]), (group_data[from_name][3], group_data[to_name][3]), color=color, linestyle=linestyle)

plt.legend()
plt.xlabel('Deepstream FPS')
plt.ylabel('mAP 0.5:0.95')
plt.show()
```
